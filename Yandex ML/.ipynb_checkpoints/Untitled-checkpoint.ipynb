```python
import pandas as pd
from sklearn.ensemble import RandomForestRegressor
from sklearn.ensemble import GradientBoostingRegressor
def roman_to_int(s):
    roman = {'I':1, 'II':2, 'III':3, 'IV':4, 'V':5,
             'VI':6, 'VII':7, 'VIII':8, 'IX':9, 'X':10}
    return roman.get(s, 0)


train = pd.read_csv('train.csv')
test = pd.read_csv('test.csv').reset_index() 


if 'id' not in test.columns:
    test = test.rename(columns={'index': 'id'})


test['B'] = test['B'].apply(roman_to_int).astype(float)
train['B'] = train['B'].apply(roman_to_int).astype(float)


#model = RandomForestRegressor(n_estimators=300, random_state=76)
model = GradientBoostingRegressor(
    n_estimators=500,
    learning_rate=0.05,
    max_depth=7,
    subsample=0.8,
    random_state=76)
model.fit(train[['A','B','C','D','E','F']], train['target'])


result = pd.DataFrame({
    'id': test.index, 
    'target': model.predict(test[['A','B','C','D','E','F']])
})

result.to_csv('answers.csv', index=False)
```

```python
import pandas as pd
import numpy as np
from sklearn.ensemble import GradientBoostingRegressor

def preprocess_features(df):
    
    if 'id' not in df.columns:
        df['id'] = df.index.astype(int) + 1  # +1 чтобы начиналось с 1
    
    
    roman_dict = {'I':1, 'II':2, 'III':3, 'IV':4, 'V':5}
    for col in ['A','B','C','D','E','F']:
        
        if df[col].dtype == object:
            df[col] = df[col].apply(lambda x: roman_dict.get(str(x).strip(), np.nan))
        # Заполняем пропуски
        df[col].fillna(df[col].median(), inplace=True)
    
    return df


train = pd.read_csv('train.csv')
test = pd.read_csv('test.csv')

train = preprocess_features(train)
test = preprocess_features(test)


if 'id' not in test.columns:
    test['id'] = test.index.astype(int) + 1


model = GradientBoostingRegressor()
model.fit(train[['A','B','C','D','E','F']], train['target'])

result = pd.DataFrame({
    'id': test['id'],
    'target': model.predict(test[['A','B','C','D','E','F']])
})

result.to_csv('answers.csv', index=False)
```
